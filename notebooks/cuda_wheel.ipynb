## Build Wheel File
- Could take 15 - 20 mins

```python
import os
os.environ["CMAKE_ARGS"] = "-DGGML_CUDA=ON -DGGML_NATIVE=OFF -DLLAMA_NATIVE=OFF"
os.environ["CMAKE_BUILD_PARALLEL_LEVEL"] = "4"

!pip wheel "llama-cpp-python>=0.3.0" \
    --no-binary llama-cpp-python \
    --wheel-dir /content/wheels \
    --no-deps
```
